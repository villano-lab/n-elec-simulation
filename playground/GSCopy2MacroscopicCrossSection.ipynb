```python
#we may need some code in the ../python directory and/or matplotlib styles
import sys
sys.path.append('../python/')

#matplotlib for plotting
import matplotlib as mpl
from matplotlib import pyplot as plt
plt.style.use('../mplstyles/stylelib/standard.mplstyle')

import numpy as np
```

```python
#Calculations for number density and other values
#Want microscopic cross section s where 
s=8.264e-24 #Elastic cross section of lead 209 at 10keV in cm^2
print(s)
rho=11.34 #in g/cm^3
m=3.469e-22 #in g
n=rho/m
print(n)
```

```python
sigma=n*s
print(sigma)
```

```python
a = np.random.exponential(scale=sigma, size=1000)
print(a)
```

```python
ebins = np.linspace(0,5,100)
xcq = (ebins[:-1] + ebins[1:]) / 2
n,nx = np.histogram(a,bins=ebins)

#nel,nxel = np.histogram(a,bins=ebins)
```

```python
#set up a 1d plot
fig,axes = plt.subplots(1,1,figsize=(15.0,8.0),sharex=True)
ax1 = axes


#thist = n/np.sum(n)/np.diff(xcq*1000)[0]


#print(estring)
ax1.step(xcq,n, where='mid',color='m', linestyle='-', \
            label='Exponential', linewidth=2)


ymin = 0
ymax = 200

blue = '#118DFA'
ax1.fill_between(xcq,np.zeros(np.shape(xcq)),n,step='mid',facecolor='b',alpha=0.4, \
                 label='')


#ax1.set_yscale('linear')
ax1.set_yscale('linear')
ax1.set_xlim(0.0, 5) 
ax1.set_ylim(ymin,ymax)
ax1.set_xlabel(r'value')
ax1.set_ylabel('counts')
ax1.grid(True)
ax1.yaxis.grid(True,which='minor',linestyle='--')
ax1.legend(loc=1,prop={'size':22})
#ax1.legend(bbox_to_anchor=(1.04,1),borderaxespad=0,prop={'size':22})

for axis in ['top','bottom','left','right']:
  ax1.spines[axis].set_linewidth(2)

plt.tight_layout()
#plt.savefig('paper_figures/hist.eps')
plt.savefig('hist.pdf')
plt.show()
```

```python
#Find elastic cross sections
sc=4.95e-24 #Elastic cross section of Carbon 12 at 10keV in cm^2
rhoc=2.26 #in g/cm^3
mc=1.99269e-23 #in g
nc=rhoc/mc
sigmac=nc*sc
print(sigmac)

sh=3.027e-23 #Elastic cross section of Hydrogen 1 at 10keV in cm^2
rhoh=0.0000899 #in g/cm^3
mh=1.67e-24 #in g
nh=rhoh/mh
sigmah=nh*sh
print(sigmah)

su=9.24e-24 #Elastic cross section of Uranium 238 at 10keV in cm^2
rhou=19.05 #in g/cm^3
mu=3.9521e-22 #in g
nu=rhou/mu
sigmau=nu*su
print(sigmau)

sl=s #Elastic cross section of Uranium 238 at 10keV in cm^2
rhol=rho #in g/cm^3
ml=m #in g
nl=rhol/ml
sigmal=nl*sl
print(sigmal)

```

```python
#Number lists
ac = np.random.exponential(scale=sigmac, size=10000)
ah = np.random.exponential(scale=sigmah, size=10000)
au = np.random.exponential(scale=sigmau, size=10000)
al = np.random.exponential(scale=sigmal, size=10000)
```

```python
#histograms
ebinsnew = np.linspace(0,0.1,100)
xcqnew = (ebinsnew[:-1] + ebins[1:]) / 2


nc,nx = np.histogram(ac,bins=ebinsnew)
nh,nx = np.histogram(ah,bins=ebinsnew)
nu,nx = np.histogram(au,bins=ebinsnew)
nl,nx = np.histogram(al,bins=ebinsnew)
```

```python
#set up a 1d plot
fig,axes = plt.subplots(1,1,figsize=(15.0,8.0),sharex=True)
ax1 = axes


#thist = n/np.sum(n)/np.diff(xcq*1000)[0]


#print(estring)
ax1.step(xcqnew,nl, where='mid',color='m', linestyle='-', \
            label='Lead209', linewidth=2)
ax1.step(xcqnew,nc, where='mid',color='k', linestyle='-', \
            label='Carbon12', linewidth=2)
ax1.step(xcqnew,nh, where='mid',color='r', linestyle='-', \
            label='Hydrogen1', linewidth=2)
ax1.step(xcqnew,nu, where='mid',color='b', linestyle='-', \
            label='Uranium238', linewidth=2)

ymin = 0
ymax = 5000


#ax1.set_yscale('linear')
ax1.set_yscale('linear')
ax1.set_xlim(0.0, 5)
#changed xlim away from 5 to set scale for H-1
ax1.set_ylim(ymin,ymax)
ax1.set_xlabel(r'value')
ax1.set_ylabel('counts')
ax1.grid(True)
ax1.yaxis.grid(True,which='minor',linestyle='--')
ax1.legend(loc=1,prop={'size':22})
#ax1.legend(bbox_to_anchor=(1.04,1),borderaxespad=0,prop={'size':22})

for axis in ['top','bottom','left','right']:
  ax1.spines[axis].set_linewidth(2)

plt.tight_layout()
#plt.savefig('paper_figures/hist.eps')
#plt.savefig('hist.pdf')
plt.show()
```
